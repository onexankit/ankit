#step 1:import liabrary

```python
import pandas as pd
```

#step 2:import data salary = pd.read_csv('')

```python
salary = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
salary.head()
```

```python
salary.info()
```

#step 3:define target(Y) and feature (X)

```python
salary.columns
```

```python
y = salary['Salary']
x = salary[['Experience Years']]
```

```python
salary.shape
```

```python
x.shape
```

```python
y.shape
```

#step 4:train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
x_train
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# step 5:select model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

# step 6:train model (fit model)   machine learning happens

```python
model.fit(x_train, y_train,)
```

```python
model.intercept_
```

```python
model.coef_
```

#step 7:prediction

```python
y_pred = model.predict(x_test)
```

```python
x_test
```

#step 8:accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
