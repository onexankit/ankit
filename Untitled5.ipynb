```python
#step 1: import liabrary
import pandas as pd
```

```python
#step 2: import data
salary = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Salary%20Data.csv')
```

```python
 salary.head()
```

```python
 salary.info()
```

```python
salary.describe()
```

```python
#step 3: define target (y) and feature (x)
salary.columns
```

```python
y = salary['Salary']
x = salary[['Experience Years']]
```

#step 4: split

```python
 from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,random_state=2529)
```

# step 5: select model

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

# step 6: train model (fit model) machine learing happens here

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

# step 7: prediction

```python
y_pred = model.predict(x_test)
```

```python
x_test
```

# step 8: accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error (y_test,y_pred)
```
